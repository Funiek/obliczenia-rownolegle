```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd
import numpy as np
from matplotlib.tri import Triangulation
from matplotlib.ticker import ScalarFormatter
```

```python
df = pd.read_csv('wyniki.csv')
df
```

```python
x = df['Ilość bajtów']
y = df['Ilość cykli']
z = df['Czas trwania wszystkich operacji - Blokujące']

# Inicjalizuj wykres 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Ustal etykiety osi
ax.set_xlabel('Ilość bajtów')
ax.set_ylabel('Ilość cykli')
ax.set_zlabel('Czas trwania wszystkich operacji - Blokujące')
#ax.set_xticks(np.arange(0, 32769, 128))
#ax.set_yticks(np.arange(0, 32769, 128))

#ax.set_xlim(0, 32768)  # Zakres dla osi X
#ax.set_ylim(0, 32768)  # Zakres dla osi Y

# Tworzenie triangulacji punktów
tri = Triangulation(x, y)

# Utwórz wykres 3D z wypełnieniem między punktami
ax.plot_trisurf(x, y, z, triangles=tri.triangles, cmap='viridis', linewidth=0.2)
plt.savefig('Czas trwania wszystkich operacji - Blokujące.png', dpi=300)
# Wyświetl wykres

#ax.view_init(30, -90) 
plt.show()
```

```python
p_df = df[df['Ilość bajtów']==1]
p_df
```

```python
# Dane punktów dla pierwszej linii
x1 = p_df['Ilość cykli']
y1 = p_df['Czas trwania wszystkich operacji - Blokujące']

# Dane punktów dla drugiej linii
x2 = p_df['Ilość cykli']
y2 = p_df['Czas trwania wszystkich operacji - Nieblokujące']

# Tworzenie wykresu liniowego dla pierwszej linii
plt.plot(x1, y1, linestyle='-', color='b', label='Blokujące')

# Tworzenie wykresu liniowego dla drugiej linii
plt.plot(x2, y2, linestyle='-', color='r', label='Nieblokujące')
# Dodaj etykiety osi
plt.xlabel('Ilość cykli')
plt.ylabel('Czas trwania wszystkich operacji [s]')

# Dodaj tytuł wykresu
plt.title('Czas trwania wszystkich operacji')
plt.legend()
plt.savefig('Czas trwania wszystkich operacji 2D.png', dpi=300)

# Wyświetl wykres
plt.show()
```

```python
# Dane punktów dla pierwszej linii
x1 = p_df['Ilość cykli']
y1 = p_df['Szybkość łącza - Blokujące']

# Dane punktów dla drugiej linii
x2 = p_df['Ilość cykli']
y2 = p_df['Szybkość łącza - Nieblokujące']

# Tworzenie wykresu liniowego dla pierwszej linii
plt.plot(x1, y1, linestyle='-', color='b', label='Blokujące')

# Tworzenie wykresu liniowego dla drugiej linii
plt.plot(x2, y2, linestyle='-', color='r', label='Nieblokujące')
plt.gca().xaxis.set_major_formatter(ScalarFormatter(useMathText=True))
plt.gca().yaxis.set_major_formatter(ScalarFormatter(useMathText=True))
# Dodaj etykiety osi
plt.xlabel('Ilość cykli')
plt.ylabel('Szybkość łącza')

# Dodaj tytuł wykresu
plt.title('Szybkość łącza')
plt.legend()
plt.savefig('Szybkość łącza 2D.png', dpi=300)
# Wyświetl wykres
plt.show()
```

```python
# Dane punktów dla pierwszej linii
x1 = p_df['Ilość cykli']
y1 = p_df['Czas wykonania pojedynczej operacji - Blokujące']*1_000_000

# Dane punktów dla drugiej linii
x2 = p_df['Ilość cykli']
y2 = p_df['Czas wykonania pojedynczej operacji - Nieblokujące']*1_000_000

# Tworzenie wykresu liniowego dla pierwszej linii
plt.plot(x1, y1, linestyle='-', color='b', label='Blokujące')

# Tworzenie wykresu liniowego dla drugiej linii
plt.plot(x2, y2, linestyle='-', color='r', label='Nieblokujące')

# Dodaj etykiety osi
plt.xlabel('Ilość cykli')
plt.ylabel('Czas wykonania pojedynczej operacji (µs)')

# Dodaj tytuł wykresu
plt.title('Czas wykonania pojedynczej operacji')
plt.legend()
plt.savefig('Czas wykonania pojedynczej operacji 2D.png', dpi=300)
# Wyświetl wykres
plt.show()
```
